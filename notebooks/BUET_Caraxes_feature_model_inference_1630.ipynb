```python
import pandas as pd
df = pd.read_csv(test_data_path_2,sep='\n',header=None, encoding="UTF-8")
df.head(20)
```

```python
test_data_path = test_data_path_2
```

```python
df = df.rename(columns={0: 'text'})
df
```

```python
import numpy as np
embeddings_dict = {}
glove_path = "/content/drive/MyDrive/nlp_hackathon_bd_2023-main/bn_glove.39M.100d.txt"
with open(glove_path, 'r') as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings_dict[word] = vector
```

```python
def tovec(label):
  if label in embeddings_dict:
    return embeddings_dict[label]
  return np.zeros((100,))

df['text'] = df['text'].apply(tovec)
```

```python
text_df = pd.DataFrame(df['text'].tolist())
df_concat = pd.concat([df, text_df], axis=1)
df_concat.drop(['text'],axis=1,inplace=True)
```

```python
df_concat
```

```python
from joblib import dump, load
loaded_model = load('/content/drive/MyDrive/nlp_hackathon_bd_2023-main/data/voting_classifier.joblib')
#you can use this loaded_model for prediction
y_pred=loaded_model.predict(np.array(df_concat))
y_pred
```

```python
tag_map = np.array(['O', 'B-LOC', 'B-GRP', 'I-GRP', 'B-PROD', 'B-CW', 'I-CW', 'B-CORP',
       'B-PER', 'I-PER', 'I-CORP', 'I-PROD', 'I-LOC'], dtype=object)
tag_map
```

```python
prediction2 = []
for i in range(len(y_pred)):
  # print(prediction[i])
  prediction2.append(tag_map[y_pred[i]])
prediction2
```

```python
len(prediction2)
```

```python
f2 = open('/content/drive/MyDrive/nlp_hackathon_bd_2023-main/data/test_labels_pred_file.txt','w')

count_enters = 0
count=0
idx=0
with open(test_data_path, 'r') as f:
    for line in f:
      if len(line)<2:
        f2.write('\n')
        count_enters+=1
      else:
        if count == lf-1:
          f2.write(prediction2[idx])
          break
        f2.write(prediction2[idx]+"\n")
        idx+=1
      count+=1
f2.close()
print(count_enters)
print(count)
```
